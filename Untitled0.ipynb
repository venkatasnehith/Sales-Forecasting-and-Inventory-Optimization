 Forecasting Models: ARIMA

```python
# Resample to monthly or weekly before forecasting
sales_series = df['Weekly_Sales'].resample('W').sum()  # or use 'M' for monthly
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(sales_series, order=(5,1,0))  # Adjust (p,d,q) if needed
model_fit = model.fit()
forecast = model_fit.forecast(steps=12)  # Forecast next 12 weeks (or months)

# Show forecast values
print(forecast)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(sales_series, label='Historical Sales')
plt.plot(forecast.index, forecast.values, label='Forecast', color='red')
plt.title('Sales Forecast (ARIMA)')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Prepare historical and forecast data

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Resample to monthly
monthly_sales = df['Weekly_Sales'].resample('M').sum()

# Fit ARIMA model
model = ARIMA(monthly_sales, order=(5,1,0))
model_fit = model.fit()
forecast = model_fit.forecast(steps=6)  # forecast next 6 months
```

Plot historical + forecast

```python
plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index, monthly_sales.values, label='Historical Sales')
plt.plot(forecast.index, forecast.values, label='Forecast', color='red')
plt.title('Monthly Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

INVENTORY OPTIMIZATION

 Economic Order Quantity (EOQ)

```python
import numpy as np

# Monthly demand estimate from forecast (e.g., 6-step ARIMA monthly forecast)
D = forecast.mean()  # Forecasted monthly demand
S = 100              # Ordering cost per order (can be from business)
H = 5                # Holding cost per unit per year (example)

EOQ = np.sqrt((2 * D * S) / H)
print(" Optimal EOQ (Order Quantity):", round(EOQ, 2))
```

Reorder Point Calculation

```python
lead_time = 5  # in days
daily_demand = forecast.mean() / 30  # convert from monthly to daily
reorder_point = lead_time * daily_demand

print(" Reorder Point:", round(reorder_point, 2))
```
